```python
#to show plots, with or without using plt.show().now no longer nessary to use
%matplotlib inline
import matplotlib.pyplot as plt
```

```python

age=[23,30,35,40,45,50]
JDSalary=[100,200,30,400,100,2000]
PDSalary=[200,500,800,200,100,600]
plt.plot(age,PDSalary,label='java developers',color='k',linestyle='--',marker='o')
plt.plot(age,JDSalary,label='python developers',linewidth=2)
plt.xlabel("age")
plt.ylabel("salary in(k$)")
plt.title('salary of developers with different ages')
plt.grid=True#show agrid view
plt.xkcd()#to give different look for our plot
plt.style.available#show all available styles to customize
plt.style.use('fivethirtyeight')#one style from all available styles
plt.legend()#adding more information for every value of y axis
plt.suptitle('general information',fontsize=18,fontweight="bold", y=1.22)
plt.savefig('plot.png')
plt.show()#the final stage to show plot
```

# plots from dictionary

```python
dic={
    'almond Butter':10,
    'peanut butter':8,
    'cashew butter':7   
}
```

```python
plt.bar(dic.keys(),dic.values());
plt.xlabel("type of butter ")
plt.ylabel("price in (birr)")
plt.title("butter price bar plot")
```

```python
plt.barh(list(dic.keys()),list(dic.values()));
plt.xlabel("price in (birr) ")
plt.ylabel("type of butter")
plt.title("butter price bar plot")
```

```python
import pandas as pd
l=np.linspace(1,2,10)# 10 elements from 1 upto 2
x=[1,2,3,4]
y=[11,12,13,14]
n=np.random.randn(2000)
dic={
    'almond Butter':10,
    'peanut butter':8,
    'cashew butter':7   
}
```

```python
plt.style.use('default')#style to use
plt.grid=False                                                                      
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                     ncols=2,
                                     figsize=(10,5))
ax1.plot(x,y),ax2.bar(dic.keys(),dic.values()),ax3.hist(n),ax4.scatter(l,l**2);
```

# plotting from pandas dataframe

```python
import pandas as pd
sample=pd.read_csv('sample.csv')
sample
```

```python
sample.drop("Unnamed: 0.1",axis=1,inplace=True)
```

```python
sample
```

```python
sample["record_date"]=pd.date_range("1/1/2022",periods=len(sample))
sample
```

```python
import matplotlib.pyplot as plt
sample.plot("zone","massawa_road_duration",kind="scatter")
```

```python
#adding the horizontal line that is the mean of messawa_road_duration
plt.axhline(sample["massawa_road_duration"].mean(), color='red', linestyle='--');
```

```python
sample.plot.bar()#x axis refers index and y axis refers value
```

```python
#bar x refers index
sample["latitude"].plot.bar()# the same with sample.plot(kind="bar")
```

```python
#x axis refers inteval and y axis refers frequency
sample["longitude"].plot(kind="hist",bins=10)#default bin 10
```

```python
sample.plot.hist(figsize=(10,10),subplots=True)
```

```python
#we can apply all available styes on this graph
fig,ax=plt.subplots( figsize=(10,5))
ax.scatter(x=sample["latitude"],y=sample["massawa_road_distance"],cmap="plasma")
```
